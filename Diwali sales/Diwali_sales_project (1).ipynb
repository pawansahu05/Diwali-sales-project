#**Exploratory Data Analysis**

**Gender**

#**ploting a bar chart for gender and its count**

```python
ax=sns.countplot(x='Gender',data=a,color='green')
for bars in ax.containers:
  ax.bar_label(bars)
```

**ploting a bar chart for gender vs total amount**

```python
ss=a.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Gender',y='Amount',data=ss)
```

**From the above graphs we can see that most of the buyers are females and even the puchasing power of females are greater then man.**

**Age**

```python
ax=sns.countplot(x='Age Group',data=a,color='salmon',hue='Gender')
for bars in ax.containers:
  ax.bar_label(bars)
```

**Total amount vs age group**

```python
sales_age=a.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group',y='Amount',data=sales_age)
```

**From the above graph we can see that most the buyers are from the age group of 26-35 years .**

**States**

#Total number of orders from the top 10 states

```python
sales_age=a.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='State',y='Orders',data=sales_age,color='red')
```

**From the above graph we can see that most of the orders and total sales\amount are from uttar pradesh, maharashtra, karnataka, delhi etc.**

**Marital Status**

```python
ac=sns.countplot(data=a,x='Marital_Status',color='indigo')
sns.set(rc={'figure.figsize':(5,5)})
for bars in ac.containers:
  ac.bar_label(bars)
```

```python
sales_age=a.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(5,5)})
sns.barplot(x='Marital_Status',y='Amount',data=sales_age,hue='Gender')
```

**From the above graphs we can see observe that most of buyers are married(woman)and they have high purchasing power.**

**Occupation**

```python
sns.set(rc={'figure.figsize':(23,5)})
ax=sns.countplot(data=a,x='Occupation')
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sale_State=a.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='Occupation',y='Amount',data=sale_State)
```

```python
sales_state=a.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='Product_Category',y='Amount',data=sales_state,color='red')
for bars in ax.containers:
  ax.bar_label(bars)
```

**From the above graph we can see that most of the sold product are from the food ,clothing and electronic category.**

```python
Sales_s=a.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
barplot=sns.barplot(x='Product_ID',y='Orders',data=Sales_s,color='red')
for bars in ax.containers:
  ax.bar_label(bars)
```
